```python
import os
import numpy as np
import pandas as pd
import re
```

## File Path

```python
path=r'spectra\may 21 (canopy photosynthesis)/'
save_path=r'CsvFile/'
```

```python
files=os.listdir(path)
len(files)
```

```python
f=files[0]
file_path = os.path.join(path, f)
with open(file_path, 'r') as file:
        lines = file_path.readlines()
```

```python

di_n_w = {'wavelengths': [], 'name': []}
df_list = []

for f in files:
    file_path = os.path.join(path, f)
    with open(file_path, 'r') as file:
        lines = file.readlines() # Select line which contains data
        name=lines[0].split('\t')[1].strip().split("_")
  
        name=str(name[1]+'_'+name[2].split('.')[0][-1])
        di_n_w['name'].append(name)  # Assigning name of each text data
        
        reflectance = []
        for line in lines[1:]:
            data = line.split()  # Assume the data is separated by whitespace
            if len(data) >= 2:
                try:
                    reflectance.append(float(data[1].strip()))
                except ValueError:
                    continue  # Skip lines that do not contain valid float numbers
        
        df_list.append(pd.DataFrame(reflectance, columns=[di_n_w['name'][-1]]).T)

        if f == files[-1]:  # Extract wavelength only once from the last file
            for line in lines[1:]:
                data = line.split()
                if len(data) > 0:
                    try:
                        di_n_w['wavelengths'].append(int(data[0].strip()))
                    except ValueError:
                        continue  # Skip lines that do not contain valid int numbers

# Combining all DataFrames
df_final = pd.concat(df_list, ignore_index=True)
df_final.columns = di_n_w['wavelengths']  # Set wavelengths as column names
df_final.index = di_n_w['name']  # Set plant names as row indices

df_final.head()  # Display the first few rows of the DataFrame
```

```python
l,r=0,4 # initial values
sample=4 #window size for averaging
dic={}
for i in range(int(df_final.shape[0]//4)):
    dic[df_final.iloc[l:r].index[0][:-2]]=df_final.iloc[l:r].mean() #taking sample of window size and assigning to a dic.
    l,r=r,r+sample #updating the initial value
    
df_71=pd.DataFrame(dic).T 

print(f" Shape {df_71.shape} \n Successful")

print(f'saving file to {save_path}')
df_71.to_csv('may20_spectra.csv')
# df_final.to_csv('test_all.csv')
```
